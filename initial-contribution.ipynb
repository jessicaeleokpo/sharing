-Creating a variable to store the non object variable and storing it in a new dataframe to ensure only numeric data

```python
col = data_c.columns[data_c.dtypes != 'object']
dataC = data_c[col]
```

-Viewing the head and shape of the new dataframe

```python
print(dataC.shape)
dataC.head()
```

-Checking for null values in our new dataset

```python
dataC.isna
().sum()
```

- As we can see from the previous output na values are present in 3 columns, now we want to eliminate na values

```python
#first we replace the na values in with their respective mean we got from the data summary because it is a continuos data and we dont want to lose information

dataC['LotFrontage'].fillna(70.04, inplace = True)

dataC['MasVnrArea'].fillna(103.685, inplace = True)

dataC['GarageYrBlt'].fillna(1978, inplace = True)
```

- Now we are going to run a linear regression of each feature against the target variable ('SalePrice') to get the individual 
correlation

```python
for i in col:
    if i == 'SalePrice':
        continue
    else:
        x = dataC[i]
        y = dataC['SalePrice']

        plt.style.use(style='ggplot')

        plt.ylabel('SalePrice')
        plt.xlabel(i)

        slope, intercept, r, p, std_err = stats.linregress(x, y)

        def myfunc(x):
          return slope * x + intercept

        mymodel = list(map(myfunc, x))

        plt.scatter(x, y)
        plt.plot(x, mymodel)
        plt.show()
        print(r)
```

Conclusion:

From the linear regression demonstrated in the chart above, it shows that the field which best describes the SalesPrice field is the OverallQual field, this is so because the OverallQual has the highest Correlation which makes it our best predictor and the most effective.
